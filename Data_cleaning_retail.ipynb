```python
import kagglehub
from  kagglehub import KaggleDatasetAdapter
import pandas as pd


import os
path = r"C:\Users\Lorenzo\.cache\kagglehub\datasets\mohammadtalib786\retail-sales-dataset\versions\1"
print(os.listdir(path))


df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "mohammadtalib786/retail-sales-dataset",
    "retail_sales_dataset.csv"  # sostituzione nome
)

print("First 5 records:", df.head())
print("Shape of the dataset:", df.shape)
# valori mancanti per colonna
df.isna().sum().sort_values(ascending=False)
# percentuale missing
(df.isna().mean()*100).sort_values(ascending=False)
df = df.dropna()

df.info()

df['Date'] =  pd.to_datetime(df['Date'])
df.info()
df.dtypes

df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day_of_week'] = df['Date'].dt.day_name()
df['dow_num'] = df['Date'].dt.dayofweek

#check ricavi
(df['Quantity'] * df['Price per Unit'] == df['Total Amount']).mean() # calcola la percentuale di righe in cui la colonna 'Total Amount' Ã¨ corretta
# media prezzo per transazione
df['prezzo_medio'] = df['Total Amount'] / df['Quantity']
print("Prezzo medio per transazione:", df['prezzo_medio'].mean())

df.head()
df.describe()
df.info()
df.to_csv(r"C:/Users/Lorenzo/Desktop/Training_py/retail_sales_cleaned.csv", index=False)
```
